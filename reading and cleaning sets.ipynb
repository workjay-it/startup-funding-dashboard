# Crunchbase Venture Analysis

## Project Goal
Analyze startup funding, investor activity, and acquisition outcomes
using Python, SQL (DuckDB), and real-world Crunchbase-style data.

## Key Business Questions
1. Which companies raised the most funding?
2. Which investors are the most active?
3. How does funding amount relate to acquisition outcomes?
4. Which funding stages attract the most capital?
5. Are highly funded startups more likely to be acquired?

## Tools Used
- Python (pandas, matplotlib)
- Jupyter Notebook (VS Code)
- DuckDB (SQL analytics)
- DBeaver (SQL IDE)
